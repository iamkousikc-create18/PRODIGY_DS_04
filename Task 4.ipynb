```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from wordcloud import WordCloud,STOPWORDS
```

## Task 4: Sentiment Analysis and Visualization of Social Media Data

### Objective
To analyze sentiment patterns in social media data and understand public opinions and attitudes toward different topics or brands. The goal is to explore the distribution of positive, negative, and neutral sentiments in the Twitter dataset and visualize these patterns using various graphical techniques such as bar charts, pie charts, and word clouds.

```python
df=pd.read_csv("twitter_training.csv",header=None)
df.columns=["ID","Topic","Sentiment","Tweet"]
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df["Sentiment"].value_counts()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x="Sentiment",data=df)
plt.title("Sentiment Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x="Topic",hue="Sentiment",data=df)
plt.title("Sentiment by Topic")
plt.xticks(rotation=90)
plt.show()
```

```python
positive=df[df["Sentiment"]=="Positive"]["Tweet"]
text=" ".join(positive.astype(str))
wordcloud=WordCloud(width=800,height=400,background_color="white",stopwords=STOPWORDS).generate(text)
plt.imshow(wordcloud)
plt.axis("off")
plt.title("Positive Tweets Word Cloud")
plt.show()
```

```python
negative=df[df["Sentiment"]=="Negative"]["Tweet"]
text=" ".join(negative.astype(str))
wordcloud=WordCloud(width=800,height=400,background_color="white",stopwords=STOPWORDS).generate(text)
plt.imshow(wordcloud)
plt.axis("off")
plt.title("Negative Tweets Word Cloud")
plt.show()
```

```python
sentiment_count=df["Sentiment"].value_counts()
plt.pie(sentiment_count,labels=sentiment_count.index,autopct="%1.1f%%")
plt.title("Sentiment Percentage")
plt.show()
```

# Insights

- The dataset contains tweets expressing positive, negative, and neutral sentiments related to various topics or brands.
- The sentiment distribution visualization shows the overall public opinion trend in the dataset.
- Topic-wise sentiment analysis highlights how different topics receive different levels of positive or negative reactions.
- Word cloud visualization helps identify frequently used words in positive tweets, indicating common user opinions and expressions.
- Overall, sentiment analysis provides useful insights into public attitudes and trends in social media discussions.
